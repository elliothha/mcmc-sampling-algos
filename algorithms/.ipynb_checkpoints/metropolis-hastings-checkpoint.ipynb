```python
import torch

import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline

torch.manual_seed(1337)
```

```python
def target_dist(x):
    '''
    Target distribution = unnormalized distribution that we want to sample from
    Returns log p(x) = log probability of state x occuring under p
    '''
    return torch.log(torch.exp(x)) # Using e^x as the target dist, so log prob will just be x

def propose(x, sigma):
    '''
    Adds random noise to x to generate a new state, y, in our random walk kernel
    Returns y = the newly proposed state based on the current state
    '''
    return x + sigma * torch.randn_like(x)

def proposal_dist(y, x, sigma):
    '''
    The chosen transition kernel/proposal dist. is a Gaussian random walk proposal where we add N(0,1) to x
    This allows us to generate a new state, y, from x
    See: https://stephens999.github.io/fiveMinuteStats/MH_intro.html
    Returns log q(y|x) = log prob of state y occuring under q, given state x
    '''
    squared_diff = (y - x) ** 2
    return -0.5 * torch.log(2 * torch.pi * torch.ones_like(y)) - 0.5 * squared_diff / (sigma ** 2)

def accept_probability(y, x, sigma):
    '''
    Don't need to know the normalizing constant of the posterior dist. due to the ratio
    Returns A = the acceptance prob. based on proposed state, y, and current state, x
    '''
    log_ratio = target_dist(y) + proposal_dist(x, y, sigma) - (target_dist(x) + proposal_dist(y, x, sigma))
    exp_log_ratio = torch.exp(log_ratio)
    exp_log_ratio = torch.where(torch.isnan(exp_log_ratio), torch.tensor(1.0), exp_log_ratio)
    
    A = torch.min(
        torch.tensor(1.0),
        exp_log_ratio   
    )
    return A

def metropolis_hastings(num_iters, sigma):
    '''
    Idea is to construct a Markov chain whose stationary dist. = target dist.
    Choose proposal dist., q, s.t. Markov chain fulfills detailed balance condition
    Uses the proposal dist. to explore the state space
    Our chain converges to the stationary dist. because of ergodicity and the detailed balance condition
    '''
    samples = []
    x = torch.zeros(1) # initial state doesn't matter, can be anything
    
    for i in range(1, num_iters):
        y = propose(x, sigma) # proposed state, y, given current state, x
        A = accept_probability(y, x, sigma)
        # print(A)
        if torch.rand(1) < A:
            x = y # accept move x->y with prob. min(1, A), else keep x the same
        
        samples.append(x)
    
    return np.array([s.item() for s in samples])
```

```python
def view(samples, title):
    fig, axs = plt.subplots(2, 1, figsize=(10, 8))

    # histogram
    axs[0].hist(samples, bins=100, density=True, alpha=0.75, edgecolor='black')
    axs[0].set_title(f'Histogram of samples from $e^x$ for {title}')
    axs[0].set_xlabel('x')
    axs[0].set_ylabel('Density')

    # trace plot
    axs[1].plot(samples, marker='o', linestyle='', markersize=2)
    axs[1].set_title(f'Trace plot of the samples for {title}')
    axs[1].set_xlabel('Sample index')
    axs[1].set_ylabel('Sample value')

    plt.tight_layout()
    plt.show()
```

```python
num_iters = 100000
burnin = 10000 # burn-in values to discard
sigma = 1.0

samples = metropolis_hastings(num_iters + burnin, sigma)
view(samples[burnin:], 'Metropolis-Hastings')
```
